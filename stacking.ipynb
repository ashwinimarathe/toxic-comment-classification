```python
import pandas as pd
import numpy as np
```

```python
nb_svm = pd.read_csv('nb_svm.csv')
glove = pd.read_csv('glove_embeddings.csv')
```

```python
labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
```

```python
stacking = nb_svm.copy()
stacking[labels] = 0.5*nb_svm[labels] + 0.5*glove[labels]
```
